```python
import cv2
import numpy as np
import os
import pandas as pd
from tqdm import tqdm

def compute_mscn(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).astype(np.float32)

    # Compute the mean and standard deviation
    mean, std_dev = cv2.meanStdDev(gray_image)

    # Compute MSCN (Mean Subtracted Contrast Normalization) features
    mscn = (gray_image - mean[0][0]) / std_dev[0][0]

    return mscn

def extract_mscn_features_from_dataset(dataset_path, output_csv):
    # Create a list to store the results
    results = []

    # Loop through all the images in the dataset
    for image_name in tqdm(os.listdir(dataset_path)):
        image_path = os.path.join(dataset_path, image_name)
        image = cv2.imread(image_path)

        if image is not None:
            # Compute MSCN features
            mscn_features = compute_mscn(image).flatten()
            
            # Add the result to the list
            results.append([image_name] + mscn_features.tolist())

    # Create a DataFrame from the results
    columns = ['image_name'] + [f'mscn_{i}' for i in range(len(results[0]) - 1)]
    df = pd.DataFrame(results, columns=columns)

    # Save the DataFrame to a CSV file
    df.to_csv(output_csv, index=False)

# Path to your dataset
dataset_path = 'path_to_your_dataset'

# Path to the output CSV file
output_csv = 'mscn_features.csv'

# Extract MSCN features from the dataset and save to CSV
extract_mscn_features_from_dataset(dataset_path, output_csv)
```
